2.  Explain the basic architecture and functioning of a Perceptron. What are its limitations?
    
    - A Perceptron is the simplest type of artificial neural network used for binary classification (two-class problems). It consists of three main parts:

        - Input Layer: Receives input values (features).
        - Weights and Bias: Each input is multiplied by a weight, and a bias is added.
        - Activation Function: Produces the final output (usually 0 or 1) based on whether the weighted sum exceeds a threshold.
    
    - Working of a Perceptron:

       - The perceptron receives input values.
       - Each input is multiplied by its corresponding weight.
       - The weighted values are summed, and a bias is added.
       - The activation function decides the output:
           
           - Output = 1 if the sum is greater than the threshold.
           - Output = 0 otherwise.

       - During training, the weights are updated to reduce classification errors.

     - Limitations of a Perceptron:
        
        -It can solve only linearly separable problems.
        - It cannot solve non-linearly separable problems such as the XOR problem.
        
        - It has only a single layer, so it cannot learn complex patterns.
        - Its performance is limited for large and complex datasets.

3. Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.

     - An activation function is a mathematical function used in neural networks to determine the output of a neuron. Its main purpose is to introduce non-linearity, allowing the network to learn complex patterns and relationships in data. Without activation functions, a neural network would behave like a simple linear model and would not be able to solve complex problems.

     - Comparison of Activation Functions:

        - Sigmoid :-
             
             - Output Range - 0 to 1
             - Advantages - Good for binary classification; smooth output.
             - Disadvantages - Suffers from vanishing gradient and is slower to train.

        - Tanh (Hyperbolic Tangent) :-
             
             - Output Range - -1 to 1
             - Advantages - Zero-centered output and learns better than Sigmoid.
             - Disadvantages - Also suffers from vanishing gradient for large values.

        - ReLU (Rectified Linear Unit) :-
             
             - Output Range - 0 to ∞
             - Advantages - Fast, simple, and reduces the vanishing gradient problem.
             - Disadvantages - Neurons may stop learning if inputs are always negative ("dead ReLU").

          


4.  What is the difference between Loss function and Cost function in neural networks? Provide examples.
    
    - A Loss Function and a Cost Function are used to measure the error between the predicted output and the actual output in a neural network. The main difference is the number of training examples they consider.

    - Loss Function :-
        
        - Measures the error for a single training example.
        - Used to calculate the error for one prediction.
        - Helps compute the error for each sample.

    - Cost Function :-
         
         - Measures the average or total error over the entire training dataset.
         - Used to evaluate the overall performance of the model during training.
         - Guides the optimization algorithm to update model parameters.

    - Examples:
       
       - Loss Functions:
           
           - Mean Squared Error (MSE): Used for regression problems.
           - Binary Cross-Entropy: Used for binary classification.
           - Categorical Cross-Entropy: Used for multi-class classification.

      - Cost Function:
          
          - The average Mean Squared Error (MSE) over all training samples.
          - The average Cross-Entropy Loss over the entire dataset.
          

5. What is the role of optimizers in neural networks? Compare Gradient
Descent, Adam, and RMSprop.
   
   - An optimizer is an algorithm used to update the weights and biases of a neural network during training. Its main role is to minimize the loss (cost) function by finding the best values of the model parameters, which improves the accuracy of predictions.

   Comparison of Optimizers:
       
  - Gradient Descent :-  Updates weights using the gradient of the loss function with a fixed learning rate.
    
    - Advantages - Simple and easy to implement.

    - Disadvantages - Slow convergence and may get stuck in local minima.

  - RMSprop - Adjusts the learning rate for each parameter using recent squared gradients.

       - Advantages - Faster convergence and works well for complex problems.

       - Disadvantages - Requires tuning of hyperparameters.


  - Adam (Adaptive Moment Estimation) - Combines Momentum and RMSprop by using both past gradients and squared gradients.

     - Advantages - Fast, efficient, and widely used in deep learning.

     - Disadvantages - Uses more memory and may not always give the best generalization.

```python
# 6. Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate.

import numpy as np

# Training data for AND gate
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Target output
y = np.array([0, 0, 0, 1])

# Initialize weights and bias
weights = np.zeros(2)
bias = 0
learning_rate = 0.1
epochs = 10

# Step activation function
def activation(x):
    return 1 if x >= 0 else 0

# Training the perceptron
for epoch in range(epochs):
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        prediction = activation(linear_output)

        error = y[i] - prediction

        # Update weights and bias
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

# Testing the perceptron
print("Final Weights:", weights)
print("Final Bias:", bias)

print("\nAND Gate Predictions:")
for i in range(len(X)):
    output = activation(np.dot(X[i], weights) + bias)
    print(f"{X[i]} -> {output}")

```

```python
# Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-10, 10, 400)

# Activation Functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Compute outputs
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# Plotting
plt.figure(figsize=(15, 4))

# Sigmoid Plot
plt.subplot(1, 3, 1)
plt.plot(x, y_sigmoid, color='blue')
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)

# ReLU Plot
plt.subplot(1, 3, 2)
plt.plot(x, y_relu, color='green')
plt.title("ReLU Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)

# Tanh Plot
plt.subplot(1, 3, 3)
plt.plot(x, y_tanh, color='red')
plt.title("Tanh Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)

plt.tight_layout()
plt.show()

```

```python
# 8.  Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy.

# Import required libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist

# Load the MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize the pixel values (0-255 -> 0-1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# Build the Multilayer Neural Network
model = Sequential([
    Flatten(input_shape=(28, 28)),      # Convert 28x28 image into 784 features
    Dense(128, activation='relu'),       # Hidden layer with 128 neurons
    Dense(64, activation='relu'),        # Hidden layer with 64 neurons
    Dense(10, activation='softmax')      # Output layer (10 classes)
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=32,
    verbose=1
)

# Print Training Accuracy
train_loss, train_accuracy = model.evaluate(X_train, y_train, verbose=0)
print("Training Accuracy:", train_accuracy)

# Print Test Accuracy (Optional)
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
print("Test Accuracy:", test_accuracy)

```

```python
# 9. Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior.

# Import libraries
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import fashion_mnist

# Load Fashion MNIST dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Normalize pixel values
X_train = X_train / 255.0
X_test = X_test / 255.0

# Build the neural network
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = model.fit(
    X_train,
    y_train,
    epochs=10,
    validation_data=(X_test, y_test),
    batch_size=32
)

# Plot Loss Curve
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)

# Plot Accuracy Curve
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title("Accuracy Curve")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

```

- Case Study: Fraud Detection in Banking using Deep Learning
   
   - Designing the Deep Learning Model
      
       - Since the dataset contains structured (tabular) data, a Multilayer Neural Network (MLP) is more suitable than a single-layer perceptron because fraud detection is a complex, non-linear classification problem.

    - Proposed Model:
        
        - Input Layer: Takes features such as transaction amount, merchant ID, customer location, transaction time, etc.

        - Hidden Layer 1: 128 neurons with ReLU activation.

        - Hidden Layer 2: 64 neurons with ReLU activation.

        - Output Layer: 1 neuron with Sigmoid activation to classify transactions as Fraud (1) or Legitimate (0).

    - Activation Function and Loss Function:
    
    - Activation Function:
        
        - ReLU (Rectified Linear Unit) for hidden layers because it is fast, reduces the vanishing gradient problem, and performs well in deep neural networks.

        - Sigmoid for the output layer because fraud detection is a binary classification problem, and Sigmoid outputs a probability between 0 and 1.
  - Loss Function:
        
        - Binary Cross-Entropy Loss
        - It is the standard loss function for binary classification and measures the difference between predicted probabilities and actual labels.

  - Training and Evaluation (Considering Class Imbalance)
     
     - Since fraudulent transactions are much fewer than legitimate ones, the dataset is imbalanced.

  - Training Strategy:
      - Normalize or standardize numerical features.
      - Split the dataset into Training, Validation, and Testing sets.
      - Handle class imbalance using:
         
         - Class Weights
         - Oversampling (e.g., SMOTE)
         - Undersampling the majority class
  - Evaluation Metrics:
      
  - Instead of relying only on accuracy, use:
      
      - Precision: Measures how many predicted frauds are actually fraud.
      
      - Recall: Measures how many actual frauds are detected.
      - F1-Score: Balances Precision and Recall.
      - ROC-AUC Score: Measures the model's ability to distinguish between fraud and legitimate transactions.

  - Optimizer and Overfitting Prevention
  - Optimizer
      
      - Adam Optimizer
          
          - Fast convergence.
          - Adaptive learning rate.
          - Performs well on large datasets.

      - Preventing Overfitting:
        
        - Use Dropout layers (e.g., Dropout = 0.5).
        - Apply Early Stopping to stop training when validation loss stops improving.
        - Use L2 Regularization to reduce model complexity.
        - Train with sufficient data and validate the model on unseen data.
